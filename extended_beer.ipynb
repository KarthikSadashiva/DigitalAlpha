```python
import pandas as pd # Pandas is used for reading data from excel
import matplotlib.pyplot as plt # Matplotlib helps us in visualizing the outcomes
import numpy as np # Numpy has useful libraries for statistics

#Beer Dataset
df = pd.read_excel('C:\\Users\\user\\.spyder-py3\\beer.xlsx', sheet_name='Sheet1')

#Basic details about the data
#print(df.describe())

p12 = df['PRICE 12PK'] 
p18 = df['PRICE 18PK']
p30 = df['PRICE 30PK']
s12 = df['CASES 12PK']
s18 = df['CASES 18PK']
s30 = df['CASES 30PK']

print()
print("Prediction of pack 12 data.")
p12_train = p12[:26]
p12_test = p12[26:]

s12_train = s12[:26]
s12_test = s12[26:]

p12_slope_offset = np.polyfit(p12_train,s12_train,1)
p12_slope_offset2 = np.polyfit(p12_test,s12_test,1)

s12_predicted = np.polyval(p12_slope_offset,p12_test)
s12_predicted2 = np.polyval(p12_slope_offset2,p12_train)

plt.figure(1)
plt.plot(list(range(0,26)),s12_predicted2,marker='o')
plt.plot(list(range(0,26)),s12_train,marker='^')
plt.title("Predicted Sales of week 1 to week 26")
plt.xlabel("week")
plt.ylabel("Sales")
plt.legend(('predicted', 'actual'))
plt.show()
print("Slope =",p12_slope_offset[0]," Intercept =",p12_slope_offset[1])
mse_p12_1 = np.mean((s12_train - s12_predicted2)**2)
print("Mean Squared Error :",mse_p12_1)

plt.figure(2)
plt.plot(list(range(26,52)),s12_predicted,marker='o')
plt.plot(list(range(26,52)),s12_test,marker='^')
plt.title("Predicted Sales of week 27 to week 52")
plt.xlabel("week")
plt.ylabel("Sales")
plt.legend(('predicted', 'actual'))
plt.show()
print("Slope =",p12_slope_offset2[0]," Intercept =",p12_slope_offset2[1])
mse_p12_2 = np.mean((s12_test - s12_predicted)**2)
print("Mean Squared Error :",mse_p12_2)

print()
print("Prediction of pack 18 data.")
p18_train = p18[:26]
p18_test = p18[26:]

s18_train = s18[:26]
s18_test = s18[26:]

p18_slope_offset = np.polyfit(p18_train,s18_train,1)
p18_slope_offset2 = np.polyfit(p18_test,s18_test,1)

s18_predicted = np.polyval(p18_slope_offset,p18_test)
s18_predicted2 = np.polyval(p18_slope_offset2,p18_train)

plt.figure(3)
plt.plot(list(range(0,26)),s18_predicted2,marker='o')
plt.plot(list(range(0,26)),s18_train,marker='^')
plt.title("Predicted Sales of week 1 to week 26")
plt.xlabel("week")
plt.ylabel("Sales")
plt.legend(('predicted', 'actual'))
plt.show()
print("Slope =",p18_slope_offset[0]," Intercept =",p18_slope_offset[1])
mse_p18_1 = np.mean((s18_train - s18_predicted2)**2)
print("Mean Squared Error :",mse_p18_1)

plt.figure(4)
plt.plot(list(range(26,52)),s18_predicted,marker='o')
plt.plot(list(range(26,52)),s18_test,marker='^')
plt.title("Predicted Sales of week 27 to week 52")
plt.xlabel("week")
plt.ylabel("Sales")
plt.legend(('predicted', 'actual'))
plt.show()
print("Slope =",p18_slope_offset2[0]," Intercept =",p18_slope_offset2[1])
mse_p18_2 = np.mean((s18_test - s18_predicted)**2)
print("Mean Squared Error :",mse_p18_2)

print()
print("Prediction of pack 30 data.")
p30_train = p30[:26]
p30_test = p30[26:]

s30_train = s30[:26]
s30_test = s30[26:]

p30_slope_offset = np.polyfit(p30_train,s30_train,1)
p30_slope_offset2 = np.polyfit(p30_test,s30_test,1)

s30_predicted = np.polyval(p30_slope_offset,p30_test)
s30_predicted2 = np.polyval(p30_slope_offset2,p30_train)

plt.figure(5)
plt.plot(list(range(0,26)),s30_predicted2,marker='o')
plt.plot(list(range(0,26)),s30_train,marker='^')
plt.title("Predicted Sales of week 1 to week 26")
plt.xlabel("week")
plt.ylabel("Sales")
plt.legend(('predicted', 'actual'))
plt.show()
print("Slope =",p30_slope_offset[0]," Intercept =",p30_slope_offset[1])
mse_p30_1 = np.mean((s30_train - s30_predicted2)**2)
print("Mean Squared Error :",mse_p30_1)

plt.figure(6)
plt.plot(list(range(26,52)),s30_predicted,marker='o')
plt.plot(list(range(26,52)),s30_test,marker='^')
plt.title("Predicted Sales of week 27 to week 52")
plt.xlabel("week")
plt.ylabel("Sales")
plt.legend(('predicted', 'actual'))
plt.show()
print("Slope =",p30_slope_offset2[0]," Intercept =",p30_slope_offset2[1])
mse_p30_2 = np.mean((s30_test - s30_predicted)**2)
print("Mean Squared Error :",mse_p30_2)
```
