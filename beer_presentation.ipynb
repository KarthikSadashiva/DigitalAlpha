```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy
import seaborn as sns 
df = pd.read_excel('C:\\Users\\user\\.spyder-py3\\beer.xlsx', sheetname='Sheet1')
print(df.describe())
print("Column headings:")
#print(df.columns)
% matplotlib inline
p12 = df['PRICE 12PK']
p18 = df['PRICE 18PK']
p30 = df['PRICE 30PK']

c12 = df['CASES 12PK']
c18 = df['CASES 18PK']
c30 = df['CASES 30PK']

#Analysis of the data
plt.figure(1)
plt.subplot(321)
plt.plot(range(len(p12)),p12,'g')
plt.subplot(323)
plt.plot(range(len(p12)),p18,'r')
plt.subplot(325)
plt.plot(range(len(p12)),p30,'b')
plt.subplot(322)
plt.plot(range(len(c12)),c12,'g')
plt.subplot(324)
plt.plot(range(len(c12)),c18,'r')
plt.subplot(326)
plt.plot(range(len(c12)),c30,'b')
plt.show()
```
